```python
import pandas as pd
import pandas_profiling

# Source of file: https://data.nasa.gov/Space-Science/Meteorite-Landings/gh4g-9sfh
df=pd.read_csv("../Meteorite_Landings.csv", parse_dates=['year'], encoding='UTF-8')

#Note: Pandas does not support dates before 1880, so we ignore these for this analysis
df['year'] = pd.to_datetime(df['year'], errors='coerce')
df['source'] = "NASA"
```

```python
pandas_profiling.ProfileReport(df)
```

```python
# Optional: save profile report in HTML format
pfr = pandas_profiling.ProfileReport(df)
pfr.to_file("/tmp/example.html")
```
